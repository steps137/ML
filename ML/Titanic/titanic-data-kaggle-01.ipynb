```python
import os
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

trn_df = pd.read_csv("../input/spaceship-titanic/train.csv")
tst_df = pd.read_csv("../input/spaceship-titanic/test.csv")

def plot_float(df):
    plt.figure(figsize=(20,8), facecolor ='w')       
    cnt = 1
    for col in df:
        if df[col].dtype == np.float64:            
            plt.subplot(2, 5, cnt); plt.hist(df[col], bins=50)                            
            plt.title(col)
            cnt += 1
    plt.show()

def show_info(df):
    print("***** shape: ", df.shape)
    display(df.head(10))
    display(df.describe(include='all'))
    display(df.info())
    print("\n***** Data types:");  display(df.dtypes)    
    plot_float(df)
    print("\n***** Empty data:");  display(df.isnull().sum())
    
def fill_gaps(df):
    """ Заполняем пропуски в данных """
    for col in df:
        if df[col].dtype == np.float64:                            
            df[col] = df[col].apply(lambda x: min(x, 500) if x is not None else None)
            df.loc[df[col].isnull(), col] = df[col].median()            
        else:
            df.loc[df[col].isnull(), col] = df[col].mode()[0]    # Most frequent value    
            
            
show_info(trn_df)    

fill_gaps(trn_df)
fill_gaps(tst_df)

print("\n***** Data types:");  display(trn_df.dtypes) 
```

```python
def prepare_data(df):
    df.CryoSleep   = df.CryoSleep.astype(float)
    df.VIP         = df.VIP.astype(float)    
    
    if 'Transported' in df:
        df.Transported = df.Transported.astype(int)
    
    df.Age /= 100
    
    df['Deck']    = df.Cabin.apply(lambda x: x[0])
    df['Persons'] = df.Cabin.apply(lambda x: 0. if x[-1]=='S' else 1. )
    df['Number']  = df.Cabin.apply(lambda x: float(x[2:-2])/1000)
    df['Deck_OK'] = df.Deck.apply(lambda x: 1 if x in 'BC' else ( 0.5 if x in 'AG' else 0) )
    
    df['Check']   = df.RoomService + df.FoodCourt + df.ShoppingMall + df.Spa + df.VRDeck
    for col in  ['RoomService', 'FoodCourt',  'ShoppingMall', 'Spa', 'VRDeck']:
        #df[col] = df[col] / df['Check']
        #df.loc[df[col].isnull(), col] = 0
        df[col] /= 2000 
    df.Check  /= 2000
    
    df['FN']   = df.Name.apply(lambda x: x.split(' ')[0])
    df['SN']   = df.Name.apply(lambda x: x.split(' ')[-1])
    df['FNSX'] = df.FN.apply(lambda x: x[-1:])
    df['FN_OK']= df.FN.apply(lambda x: 0 if x[-1] in "elpy" else ( 0.5 if x[-1] in "qdgntw" else 1) )

prepare_data(trn_df)
prepare_data(tst_df)
    
display(trn_df.head())
display(trn_df[[ 'Deck_OK', 'Number']].describe())
print(trn_df.Deck.value_counts())
print(trn_df.Persons.value_counts())
print(trn_df.Number.value_counts())
print(trn_df.FNSX.value_counts()[:30])
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn import preprocessing

ONEHOT = False
if ONEHOT:
    PLANETS = ['Earth', 'Europa', 'Mars']
    onehot = OneHotEncoder(categories = [PLANETS], handle_unknown='ignore')
    trn_df = trn_df.join( pd.DataFrame(onehot.fit_transform( trn_df[['HomePlanet']] ).toarray(), columns=PLANETS ) )
    tst_df = tst_df.join( pd.DataFrame(onehot.fit_transform( tst_df[['HomePlanet']] ).toarray(), columns=PLANETS ) )

    DESTINATION = ["TRAPPIST-1e", "55 Cancri e", "PSO J318.5-22"]
    onehot = OneHotEncoder(categories = [DESTINATION], handle_unknown='ignore')
    trn_df = trn_df.join( pd.DataFrame( onehot.fit_transform(trn_df[['Destination']] ).toarray(), columns=DESTINATION ) )
    tst_df = tst_df.join( pd.DataFrame( onehot.fit_transform(tst_df[['Destination']] ).toarray(), columns=DESTINATION ) )
    
    DECKS = ['F', 'G', 'E', 'B', 'C', 'D', 'A', 'T']
    onehot = OneHotEncoder(categories = [DECKS], handle_unknown='ignore')
    trn_df = trn_df.join( pd.DataFrame(onehot.fit_transform(trn_df[['Deck']] ).toarray(), columns=DECKS ) )
    tst_df = tst_df.join( pd.DataFrame(onehot.fit_transform(tst_df[['Deck']] ).toarray(), columns=DECKS ) )
    
else:
    PLANETS = ['Planet']
    le = preprocessing.LabelEncoder()
    le.fit(trn_df.HomePlanet)
    print(le.classes_)
    trn_df = trn_df.join( pd.DataFrame( le.transform(trn_df.HomePlanet), columns=PLANETS ) )
    tst_df = tst_df.join( pd.DataFrame( le.transform(tst_df.HomePlanet), columns=PLANETS ) )    
    
    DESTINATION = ["Dest"]
    le = preprocessing.LabelEncoder()
    le.fit(trn_df.Destination)
    print(le.classes_)
    trn_df = trn_df.join( pd.DataFrame(le.transform(trn_df.Destination), columns=DESTINATION ) )
    tst_df = tst_df.join( pd.DataFrame(le.transform(tst_df.Destination), columns=DESTINATION ) )    
    
    DECKS = ['Decks']
    le = preprocessing.LabelEncoder()
    le.fit(trn_df.Deck)
    print(le.classes_)
    trn_df = trn_df.join( pd.DataFrame(le.transform(trn_df.Deck), columns=DECKS ) )
    tst_df = tst_df.join( pd.DataFrame(le.transform(tst_df.Deck), columns=DECKS ) )    
    

display(trn_df.head())
```
