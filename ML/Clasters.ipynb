# Clasters

```python
import time
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
NUM_CLASTERS = 5
x, clusters = make_blobs(n_samples = 2000, n_features = 2, 
                         centers = NUM_CLASTERS,                          #
                         cluster_std = 0.8, 
                         shuffle = True,
                         random_state=1
                        )
print(x.shape, clasters.shape)


kmeans = KMeans(n_clusters=NUM_CLASTERS, random_state=0).fit(x)
clasters = kmeans.cluster_centers_
res = kmeans.predict(x)
print(res.shape, kmeans.inertia_)



plt.figure(figsize=(8,8), facecolor ='w')
ax = plt.subplot(1,1,1)
plt.scatter(x[:,0], x[:,1], s=2, c=res)
plt.scatter(clasters[:,0], clasters[:,1], s=10, c="r")
for i in range(NUM_CLASTERS):        
    sx, sy = x[res==i].std(axis=0)    
    ax.add_patch( Ellipse(clasters[i], 2*sx, 2*sy,  color='r', fill=False) )    
plt.show()
```

```python
plt.figure(figsize=(8,5), facecolor ='w')
vars = []
for k in range(1,8):
    km = KMeans(n_clusters=k)
    km = km.fit(x)
    vars.append(km.inertia_)
    
plt.plot(np.arange(1,len(vars)+1), np.array(vars)**0.5, "-ob")
plt.axvline(5, linestyle='--', color='r')
```
