```python
import torch
import torch.nn as nn
from torch.distributions.categorical import Categorical
```

```python
#y = torch.tensor([-1., 1.] )
y = torch.tensor([[1., -1.],  [0., 10]] )


categ = Categorical(logits = y)
print(categ.sample())


print(categ.probs)
print(torch.softmax(y.view(-1,2), 1))

print()
print(categ.log_prob( torch.tensor([[0, 1], [0,1]]) ))
print(torch.log(categ.probs))
```
